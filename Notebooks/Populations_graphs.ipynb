```python
import numpy as np
import pandas as pd 
pd.set_option("display.max_columns", None)
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_pop = pd.read_excel('Populations.xlsx')

display(df_pop)
```

```python
df_pop = pd.DataFrame(df_pop)
```

```python
df_pop.columns
```

```python
plt.figure(figsize=(35,15))
sns.barplot(x='City ', y='Population',data=df_pop)
plt.xlabel('City', fontsize=17)
plt.ylabel('Population', fontsize=17)
plt.title("Populations of Cities with Clubs Competing in 2021/2022 Champions League", fontsize=20)
plt.savefig("populations.png", dpi=300)
plt.show()
```

```python
q1, q2, q3 = np.percentile(df_pop['Population'], [25, 50, 75])

print("Quartile 1 (Q1):", q1)
print("Quartile 2 (Q2):", q2)
print("Quartile 3 (Q3):", q3)
```

```python
mean = np.mean(df_pop['Population'])

print("The mean population of every competing city is:", mean)
```

```python
median = np.median(df_pop['Population'])

print("The median population of every competing city is:", median)
```

```python
stddev = np.std(df_pop['Population'])

print("Standard Deviation:", stddev)
```
